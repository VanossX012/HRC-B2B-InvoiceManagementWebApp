Filtering depending upon correlation matrix value

    We have created a function called handling correlation which is going to return fields based on the correlation matrix value with a threshold of 0.9



```python
# with the following function we can select highly correlated features
# it will remove the first feature that is correlated with anything other feature


def correlation(dataset, threshold):
    col_corr = set()  # Set of all the names of correlated columns
    corr_matrix = dataset.corr()
    for i in range(len(corr_matrix.columns)):
        for j in range(i):
            if abs(corr_matrix.iloc[i, j]) > threshold: 
                colname = corr_matrix.columns[i]  # getting the name of column
                col_corr.add(colname)
    return col_corr
```

```python
train=X_train.copy()
corr_feats=correlation(train.copy(), threshold=0.90)
corr_feats
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.tree import DecisionTreeRegressor
from sklearn.ensemble import RandomForestRegressor
from sklearn.ensemble import AdaBoostRegressor
from xgboost import XGBRegressor
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score
import pandas as pd



# Create a list of models
models = [
    LinearRegression(),
    RandomForestRegressor(),
    DecisionTreeRegressor(),
    AdaBoostRegressor(),
    XGBRegressor()
]
# NOTE: Here we didn't perform SVM since it wasn't suitable for such a large dataset, and it was taking a lot of time to execute.

# Create a list to store the results
results = []

# Function to calculate metrics and store results
def evaluate_model(model, name):
    model.fit(X_train, y_train)
    y_train_pred = model.predict(X_train)  # Predict on the training set
    y_pred = model.predict(X_test)
    train_mse = mean_squared_error(y_train, y_train_pred)
    train_rmse = mean_squared_error(y_train, y_train_pred, squared=False)
    train_r2 = r2_score(y_train, y_train_pred)
    mse = mean_squared_error(y_test, y_pred)
    rmse = mean_squared_error(y_test, y_pred, squared=False)
    r2 = r2_score(y_test, y_pred)
    results.append([name, train_mse, train_rmse, train_r2, mse, rmse, r2])

# Convert y_train and y_test to numeric type
y_train = pd.to_numeric(y_train)
y_test = pd.to_numeric(y_test)

# Evaluate each model and store results
for model in models:
    name = model.__class__.__name__
    evaluate_model(model, name)

# Create a dataframe to display the results
results_df = pd.DataFrame(results, columns=['Model', 'Train MSE', 'Train RMSE', 'Train R2 Score', 'Test MSE', 'Test RMSE', 'Test R2 Score'])

# Display the results in tabular format
results_df
```

6. Hyperparameter Tuning

        GridSearchCV was employed to select the best features/parameters 

        N_estimators: The number of decision trees being built in the forest

        Max_depth: The maximum levels allowed in a decision tree.

        

```python
from sklearn.model_selection import GridSearchCV
param_grid = {
    'n_estimators': [100, 200],
    'learning_rate': [0.1, 0.01],
    'max_depth': [3, 5],
}

xgb_regressor = xgboost.XGBRegressor()
grid_search = GridSearchCV(estimator=xgb_regressor, param_grid=param_grid, cv=5)
grid_search.fit(X_train, y_train)

best_xgb_model = grid_search.best_estimator_
best_params = grid_search.best_params_

best_xgb_predictions = best_xgb_model.predict(X_test)
best_xgb_rmse = mean_squared_error(y_test, best_xgb_predictions, squared=False)
best_xgb_r2 = r2_score(y_test, best_xgb_predictions)

print("Best XGBoost Model Evaluation:")
print("Root Mean Squared Error:", best_xgb_rmse)
print("R-squared Score:", best_xgb_r2)
```
